```julia
using Plots
```

```julia
v = [.9, .9, 1.1, 1.1, 1.5, 1.9, 1.9, 2.1, 2.1, 2.5, 2.9, 2.9, 3.1, 3.1]
```

## Histogram as a plot

```julia
h = histogram(v, bins=.5:3.5, bar_width=.9) 
# bins is a required keyword if you wish to specify
# impossible to get the underlying data when plotting
```

## Histogram using StatsBase

```julia
using StatsBase
```

```julia
hstats = fit(Histogram,v,.5:3.5)
# Can NOT use keyword bins 
# it's called edges not bins in StatsBase
```

```julia
plot(hstats, legend=false, bar_width=.9)
```

## Normalized Histograms

```julia
v = randn(10000)
```

```julia
histogram(v, normalize=true, legend=false)
```

```julia
hstat = normalize(fit(Histogram,v))  
# can't go normalize = true like plots
# normalize requires LinearAlgebra
```

```julia
using LinearAlgebra
```

```julia
hstat = normalize(fit(Histogram,v)) 
```

```julia
plot(hstat, legend=false)
```
